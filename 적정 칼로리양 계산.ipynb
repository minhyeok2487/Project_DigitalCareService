```python
import pandas as pd
```

**성인여자**

BMI 18.5 이하 = BMI x 45 ~ BMI x 59

BMI 18.5 ~ 25 = BMI x 41 ~ BMI x 57

BMI 25 ~ 30 = BMI x 38 ~ BMI x 52

BMI 30 ~ = BMI x 38.5 ~ BMi x 50

**성인남자**

BMI 18.5 이하 = BMI x 48.5 ~ BMI x 60

BMI 18.5 ~ 25 = BMI x 42 ~ BMI x 58

BMI 25 ~ 30 = BMI x 40 ~ BMI x 56

BMI 30 ~ = BMI x 41 ~ BMi x 50

**영양소**

탄수화물 = 적정 칼로리양*0.65

단백질 = 적정 칼로리양*0.15

지방 = 적정 칼로리양*0.2

콜레스테롤 섭취량 200mg

식이섬유 섭취량: 여성 20g 남성 25g

나트륨 2000mg

```python
df = pd.read_csv("현재 환자_질병데이터.csv")
df = df.loc[df["Patient Name"] == '석민규']
df = df.loc[df["검사날짜"] == '2020-06-08']
df
```

```python
def Male_cal(BMI):
    if BMI>=30:
        min_cal = BMI*41
        max_cal = BMI*50
    elif BMI>=25:
        min_cal = BMI*38
        max_cal = BMI*52
    elif BMI>18.5:
        min_cal = BMI*41
        max_cal = BMI*57
    else:
        min_cal = BMI*45
        max_cal = BMI*59
        
    min_carbohydrate = min_cal*0.65
    max_carbohydrate = max_cal*0.65
    
    min_protein = min_cal*0.15
    max_protein = max_cal*0.15
    
    min_fat = min_cal*0.2
    max_fat = max_cal*0.2
    
    Cholesterol = 200
    
    dietary_fiber = 25
    
    Sodium = 2000
    
    return (min_cal, max_cal, min_carbohydrate, max_carbohydrate, min_protein, max_protein,
            min_fat, max_fat, Cholesterol, dietary_fiber, Sodium) 

        
def Female_cal(BMI):
    if BMI<=18.5:
        min_cal = BMI*45
        max_cal = BMI*59
    elif 18.5<BMI<=25:
        min_cal = BMI*41
        max_cal = BMI*57
    elif 25<BMI<=30:
        min_cal = BMI*38
        max_cal = BMI*52
    else:
        min_cal = BMI*38.5
        max_cal = BMI*50
    return min_cal, max_cal
```

```python
BMI = float(df.loc[:,'체질량 지수'])
cal_df = pd.DataFrame(data=Male_cal(BMI),index=['최소 칼로리', '최대 칼로리', '최소 탄수화물', '최대 탄수화물',
                                               '최소 단백질', '최대 단백질', '최소 지방', '최대 지방', '콜레스테롤',
                                               '식이 섬유', '나트륨'],columns=['석민규'])
cal_df
```
