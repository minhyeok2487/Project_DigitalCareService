# 로지스틱 회귀를 이용한 질병 발생 예측
**로지스틱 회귀 : 선형 회귀 방식을 분류에 적용한 알고리즘으로**

**주로 이진 분류에 사용된다**

**필요한 모듈을 불러온후 분석에 필요한 데이터를 불러온다**

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings(action='ignore')
df = pd.read_csv("val.csv")
df.head(5)
```

**속성을 확인한다.**

**고혈압~C형간염까지의 변수들은 환자가 질병의 걸렸는지 안걸렸는지를 나타낸다**

**질병에 걸리면 1, 걸리지 않으면 0이 입력되어 있다.**

**신장~혈소판까지의 변수들은 환자 개인 특성의 데이터이다.**

**분석하기에 앞서 결측치를 찾아내서 제거한다**

```python
df=df.dropna()
df.isnull().sum()
```

### 고혈압~C형간염까지의 변수들에 대하여 회귀 분석을 진행해야 되기 때문에 필요한 작업들을 함수화한다

**로지스틱 회귀분석에서는 데이터 전처리가 필요하여 DataScaler이라는 전처리 함수를 생성하였다.**

**StandardScaler( )로 평균이 0, 분산 1로 데이터 분포를 변환한다**

```python
from sklearn.preprocessing import StandardScaler
def DataScaler(df_X):
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(df_X)
    return data_scaled
```

**학습 데이터와 테스트 데이터를 나누는 함수를 생성하였다.**

**학습 데이터를 80%, 테스트 데이터를 20%로 하였다**

```python
from sklearn.model_selection import train_test_split
def DataSplit(df_X,df_Y):
    X_train , X_test, Y_train , Y_test = train_test_split(data_scaled, df_Y, test_size=0.2, random_state=0)
    return X_train , X_test, Y_train , Y_test
```

**독립변수와 종속변수 즉, X와 Y를 분류한다**

**각종 질병 진단 여부를 종속 변수로, 나머지 변수들은 독립변수로 둔다.**

**각각의 질병 발생 예측 분석의 정확도를 알아본다.**

**그 후 환자 특성에 대한 변수들의 상관계수를 따로 추출하여 csv 파일로 저장한다.**

```python
from sklearn.metrics import accuracy_score, roc_auc_score
import statsmodels.api as sm
for i in range(26):
    df_Y = df.iloc[:, i]
    df_X = df.drop(df_Y.name,axis=1)
    data_scaled = DataScaler(df_X)
    X_train , X_test, Y_train , Y_test = DataSplit(df_X,df_Y)
    lr_clf = LogisticRegression()
    lr_clf.fit(X_train, Y_train)
    lr_preds = lr_clf.predict(X_test)
    print(df_Y.name,'의 정확도: {:0.3f}'.format(accuracy_score(Y_test, lr_preds)))
    df2 = pd.DataFrame(lr_clf.coef_)
    df2.columns=[df_X]
    df2 = df2.iloc[:,25:]
    df2.index=['{}의 상관계수'.format(df_Y.name)]
    df2.to_csv("test{}.csv".format(i))
```

### 상관계수 선정

**각 상관계수들의 상위 30%만 색깔을 다르게 하였다.**

```python
from IPython.display import Image
Image("캡처.jpg")
```

**그 중 가장 많이 중복된 변수 중 상위 10개만 선택하여 새로운 환자 개인 특성에 대한 데이터를 만들었다.**

**선택한 변수들은 아래와 같다.**

```python
dfNew = pd.read_csv("value_1.csv")
dfNew=dfNew.dropna()
print(dfNew.iloc[:,26:].columns)
Image("상관계수.jpg")
```

**선택한 변수들을 독립변수로 설정하여 질병 발생 가능성에 대한 로지스틱 회귀 예측을 하였다.**

```python
for i in range(26):
    df_Y = dfNew.iloc[:, i]
    df_X = dfNew.iloc[:,26:]
    data_scaled = DataScaler(df_X)
    X_train , X_test, Y_train , Y_test = DataSplit(df_X,df_Y)
    lr_clf = LogisticRegression()
    lr_clf.fit(X_train, Y_train)
    lr_preds = lr_clf.predict(X_test)
    print(df_Y.name,'의 정확도: {:0.3f}'.format(accuracy_score(Y_test, lr_preds)))
```

**대부분 90% 이상의 정확도를 가지는 예측 모델이 생성되었다.**

### 모델 신뢰도 평가

**예시로 고혈압과 위암의 모델 신뢰도를 알아보았다.**

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

df_Y = dfNew.iloc[:, 13]
df_X = dfNew.iloc[:,26:]
data_scaled = DataScaler(df_X)
X_train , X_test, Y_train , Y_test = DataSplit(df_X,df_Y)
lr_clf = LogisticRegression()
lr_clf.fit(X_train, Y_train)
lr_preds = lr_clf.predict(X_test)
print(df_Y.name,'의 정확도: {:0.3f}'.format(accuracy_score(Y_test, lr_preds)))

#ROC 곡선 그리기
fpr, tpr, threshold = roc_curve(Y_train, lr_clf.decision_function(X_train))

plt.plot(fpr, tpr, 'o-', ms=2, label = "Logistic Regression")
plt.legend()
plt.plot([0,1],[0,1],'k--', label="random guess")
plt.xlabel('특이도(Specificity)')
plt.ylabel('재현률(Recall)')
plt.title('위암의 ROC 커브')
plt.show()
print('위암의 AUC :{}'.format(auc(fpr, tpr)))
```

```python
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split

df_Y = dfNew.iloc[:, 0]
df_X = dfNew.iloc[:,26:]
data_scaled = DataScaler(df_X)
X_train , X_test, Y_train , Y_test = DataSplit(df_X,df_Y)
lr_clf = LogisticRegression()
lr_clf.fit(X_train, Y_train)
lr_preds = lr_clf.predict(X_test)
print(df_Y.name,'의 정확도: {:0.3f}'.format(accuracy_score(Y_test, lr_preds)))

#ROC 곡선 그리기
fpr, tpr, threshold = roc_curve(Y_train, lr_clf.decision_function(X_train))

plt.plot(fpr, tpr, 'o-',color='r', ms=2, label = "Logistic Regression")
plt.legend()
plt.plot([0,1],[0,1],'k--', label="random guess")
plt.xlabel('특이도(Specificity)')
plt.ylabel('재현률(Recall)')
plt.title('고혈압의 ROC 커브')
plt.show()
print('고혈압의 AUC :{}'.format(auc(fpr, tpr)))
```

**대부분의 모델이 70%이상의 신뢰도를 가지고 있다.**

**그 후 환자 가상의 데이터를 생성 후 위 예측모델에 적용하였다.**

**예측을 위하여 환자 개인 특성에 대한 값 들을 전처리 하였다.**

```python
df_New = pd.read_csv("현재 환자_질병데이터.csv") 
New_data_scaled = DataScaler(df_New.iloc[:,2:12])
New_data_scaled = pd.DataFrame(data=New_data_scaled,columns=df_New.iloc[:,2:12].columns)
New_data_scaled = pd.merge(df_New.iloc[:,:2],New_data_scaled,how="outer",left_index = True, right_index = True)
New_data_scaled['검사날짜'] = df_New['검사날짜']
New_data_scaled.head(5)
```

**위 로지스틱 회귀를 이용하여 한 명의 환자의 질병 발생 가능성을 예측하였다.**

**예시로 '석민규'환자의 질병 발생 가능성을 예측하였다.**

```python
for i in range(26):
    df_Y = dfNew.iloc[:, i]
    df_X = dfNew.iloc[:,26:]
    data_scaled = DataScaler(df_X)
    X_train , X_test, Y_train , Y_test = DataSplit(df_X,df_Y)
    lr_clf = LogisticRegression()
    lr_clf.fit(X_train, Y_train)
    df_New_X = New_data_scaled.loc[New_data_scaled["Patient Name"] == '석민규']
    df_New_X = df_New_X.iloc[:,2:12]
    print("{}의 발생 가능성".format(df_Y.name),lr_clf.predict(df_New_X))
```

**숫자가 0인 경우 질병이 발생하지 않을 확률이 더 큰 것이고,**

**숫자가 1인 경우 질병이 발생할 확률이 더 큰 것이다.**

**각종 질병의 발생할 확률을 알아보았다.**

```python
data = pd.DataFrame(index=["2020-06-08"])
for i in range(26):
    df_Y = dfNew.iloc[:, i]
    df_X = dfNew.iloc[:,26:]
    data_scaled = DataScaler(df_X)
    X_train , X_test, Y_train , Y_test = DataSplit(df_X,df_Y)
    lr_clf = LogisticRegression()
    lr_clf.fit(X_train, Y_train)
    df_New_X = New_data_scaled.loc[New_data_scaled["Patient Name"] == '석민규']
    df_New_X = df_New_X.loc[df_New_X["검사날짜"] == '2020-06-08']
    df_New_X = df_New_X.iloc[:,2:12]
    data["{}이 발생할 확률".format(df_Y.name)] = lr_clf.predict_proba(df_New_X)[0,1]*100
```
